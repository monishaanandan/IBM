```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
# Load the CSV files into DataFrames
df1 = pd.read_csv("drive/My Drive/IBM project/Data set/country_vaccinations_by_manufacturer.csv")
df2 = pd.read_csv("drive/My Drive/IBM project/Data set/country_vaccinations.csv")

# Merge the two DataFrames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='total_vaccinations', how='inner')
```

```python
# Save the merged DataFrame to a CSV file
merged_df.to_csv('merged_data.csv', index=False)
```
